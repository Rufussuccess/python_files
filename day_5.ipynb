```python
dict_file = [{"name": "jake", "age":30, "M": True, "Sex": ""},
            {"name": "fani", "age":20, "M": True},
             {"name": "beeniel", "age":25, "M": False},
             {"name": "susan", "age":21, "M": False}
            ]
print(dict_file)
print(type(dict_file))
```

```python
new_list = []
for a in dict_file:
    for i, j in a.items():
        if i == "age" and j >= 20:
            a["age"] = a["age"] + 5
    new_list.append(a)
print(new_list)
```

```python
import pandas as pd
df = pd.DataFrame(new_list)
df.to_csv("foolish_file2.csv", index=False)
```
