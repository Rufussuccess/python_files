```python
city_data = {
 "chicago": "C:/Users/USER/My python_folder/Data/chicago.csv",
 "new_york": "C:/Users/USER/My python_folder/Data/new_york_city.csv",
 "washington": "C:/Users/USER/My python_folder/Data/washington.csv"
}
```

```python
def get_city_data():
    city = input("enter city data: ").lower()
    df=pd.read_csv(city_data[city])
    return df

def clean_data(df):
    df[["start_date", "start_time"]] = df["Start Time"].str.split(" ", expand = True)
    df[["End_date", "End_time"]] = df["End Time"].str.split(" ", expand = True)
    df = df.drop(["Start Time", "End Time"], axis = 1)
    df["start_date"] = pd.to_datetime(df["start_date"])
    df["End_date"] = pd.to_datetime(df["End_date"])
    return df.head()
    

get_city = get_city_data()
clean_data = clean_data(get_city)
clean_data
```
